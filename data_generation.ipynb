```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
n_tasks = 300
start_dates = pd.to_datetime("2025-01-01") + pd.to_timedelta(np.random.randint(0, 60, n_tasks), unit="D")
```

```python
task_duration = np.random.randint(5, 20, n_tasks)
due_dates = start_dates + pd.to_timedelta(task_duration, unit="D")
```

```python
delays = np.random.randint(-2, 6, n_tasks)
completed_dates = due_dates + pd.to_timedelta(delays, unit="D")
```

```python
tasks = pd.DataFrame({
    "task_id": [f"T{i}" for i in range(n_tasks)],
    "start_date": start_dates,
    "due_date": due_dates,
    "completed_date": completed_dates,
    "workload_hours": np.random.randint(10, 60, n_tasks)
})

tasks.to_csv("tasks.csv", index=False)
```

```python
import random
```

```python
messages_list = [
    "Task is progressing well",
    "Blocked due to dependency",
    "Waiting for response",
    "Facing delay in implementation",
    "Need clarification",
    "Almost done",
    "No major issues"
]
```

```python
messages = []
```

```python
for task in tasks["task_id"]:
    for _ in range(np.random.randint(2, 5)):
        messages.append({
            "task_id": task,
            "sender": f"User{np.random.randint(1,6)}",
            "message": random.choice(messages_list),
            "timestamp": pd.Timestamp("2025-01-01") + pd.to_timedelta(np.random.randint(0, 60), unit="D")
        })

messages_df = pd.DataFrame(messages)
messages_df.to_csv("messages.csv", index=False)
```
